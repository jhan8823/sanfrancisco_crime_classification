```python
# Dates: 날짜의 데이터 타입을 바꾸고, 연,월,시,분,초로 나누어 새로운 칼럼을 생성
import numpy as np
# train과 test에 모두 적용
train["Dates"] = pd.to_datetime(train["Dates"])
test["Dates"] = pd.to_datetime(test["Dates"])

#train
train["Dates_year"] = train["Dates"].dt.year
train["Dates_month"] = train["Dates"].dt.month
train["Dates_day"] = train["Dates"].dt.day
train["Dates_hour"] = train["Dates"].dt.hour
train["Dates_minute"] = train["Dates"].dt.minute
train["Dates_second"] = train["Dates"].dt.second
display(train[["Dates", "Dates_year", "Dates_month", "Dates_day", "Dates_hour", "Dates_minute", "Dates_second"]].head())

#test
test["Dates_year"] = test["Dates"].dt.year
test["Dates_month"] = test["Dates"].dt.month
test["Dates_day"] = test["Dates"].dt.day
test["Dates_hour"] = test["Dates"].dt.hour
test["Dates_minute"] = test["Dates"].dt.minute
test["Dates_second"] = test["Dates"].dt.second
display(test[["Dates", "Dates_year", "Dates_month", "Dates_day", "Dates_hour", "Dates_minute", "Dates_second"]].head())
```

```python
#분을 절대적기준에 맞춰서 모음 (i.e. 5분, 55분 모두 5분 전, 후로 정리)
train['Dates_minute(abs)'] = np.abs(train['Dates_minute'] % 10)
test['Dates_minute(abs)'] = np.abs(test['Dates_minute'] % 10)
```

```python
# 평일, 주말 feature (boolean) 생성
train['Weekend'] = (train.DayOfWeek.str.contains('Saturday'))|(train.DayOfWeek.str.contains('Sunday'))
train['Weekdays'] = ~(train.DayOfWeek.str.contains('Saturday')|(train.DayOfWeek.str.contains('Sunday')))
print(train.shape)
train.head()
```

```python
#test셋에도 동일 적용
test['Weekend'] = (test.DayOfWeek.str.contains('Saturday'))|(test.DayOfWeek.str.contains('Sunday'))
test['Weekdays'] = ~(test.DayOfWeek.str.contains('Saturday')|(test.DayOfWeek.str.contains('Sunday')))

print(test.shape)
test.head()
```

```python
#PdDistrict 전처리 (one-hot encoding)
train_pddistrict = pd.get_dummies(train["PdDistrict"], prefix = "PdDistrict")

train = pd.concat([train, train_pddistrict], axis = 1)

print(train.shape)
train.head()
```

```python
test_pddistrict = pd.get_dummies(test["PdDistrict"], prefix = "PdDistrict")

# concat == concatenate
test = pd.concat([test, test_pddistrict], axis = 1)

print(test.shape)
test.head()
```

```python
# Address
#교차로 칼럼 생성 (교차로에서 발생한 사건의 경우 = True)
train["Crossroad"] = train["Address"].str.contains("/")
test["Crossroad"] = test["Address"].str.contains("/")

print(train.shape)
train[["Address", "Crossroad"]].head()
```

```python
print(test.shape)
test[["Address", "Crossroad"]].head()
```

```python
#주소 전처리 함수
def clean_address(address):
    if "/" not in address:
        return address
    
    address1,address2 = address.split("/")
    address1,address2 = address1.strip(), address2.strip()

    if address1 < address2:
        address = "{} / {}".format(address1,address2)
    else:
        address = "{} / {}".format(address2,address1)
    return address
```

```python
train['Address_clean']=train['Address'].apply(clean_address)  
train['Address_clean'].unique().shape
```

```python
test['Address_clean']=test['Address'].apply(clean_address)  
test['Address_clean'].unique().shape
```

```python
address_counts = train['Address_clean'].value_counts()
top_address_counts = address_counts[address_counts >= 100]
top_address_counts = top_address_counts.index
len(top_address_counts)
```

```python
train.loc[~train['Address_clean'].isin(top_address_counts),'Address_clean'] = 'Others'
display(train[['Address','Address_clean']].head())

test.loc[~test['Address_clean'].isin(top_address_counts),'Address_clean'] = 'Others'
display(test[['Address','Address_clean']].head())
```

```python
#one-hot 인코딩으로 주소를 전처리하기엔 메모리 소요가 크기 때문에 csr matrix를 사용해서 압축 인코딩
from scipy.sparse import csr_matrix
train_address = pd.get_dummies(train['Address_clean'])
train_address = csr_matrix(train_address)
train_address
```

```python
#coarse search
import numpy as np
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from datetime import datetime

random_state = 42

today = str(datetime.now())
filename = f"coarse-search {today}.csv"

num_loop = 100
n_estimators = 10
early_stopping_rounds = 20

coarse_hyperparameters_list = []

for loop in range(num_loop):
    num_leaves =  np.random.randint(10, 200)
    min_child_samples = np.random.randint(2, 200),
    min_child_weight = np.random.randint(0, 1.0),
    subsample = np.random.uniform(0.1, 1.0) 
    colsample_bytree = np.random.uniform(0.1, 1.0)
    learning_rate = 10 ** -np.random.uniform(low=1, high=10)
    reg_alpha = 10 ** -np.random.uniform(low=1, high=10)
    reg_lambda = 10 ** -np.random.uniform(low=1, high=15)
    drop_rate = np.random.uniform(0.1, 1.0)

    parameters ={'loop': loop,
                 'num_leaves': num_leaves, 
                 'min_child_samples': min_child_samples, 
                 'min_child_weight': min_child_weight,
                 'subsample': subsample, 
                 'colsample_bytree': colsample_bytree,
                 'reg_alpha': reg_alpha,
                 'reg_lambda': reg_lambda,
                 'n_estimators': n_estimators,
                 'learning_rate' : learning_rate,
                 'random_state': random_state}
    
    model = LGBMClassifier(**parameters)
    
    model.fit(X_train_kf, y_train_kf,
              eval_set=[(X_test_kf, y_test_kf)],
              verbose=0,
              early_stopping_rounds=early_stopping_rounds)
    
    parameters['n_estimators'] = model.best_iteration_
    parameters['score'] = model.best_score_['valid_0']['multi_logloss']
    
    print(f"{loop:2} best iteration = {parameters['n_estimators']}, Score = {parameters['score']:.6f}")
    
    coarse_hyperparameters_list.append(parameters)
    coarse_hyperparameters_data = pd.DataFrame(coarse_hyperparameters_list)
    coarse_hyperparameters_data = coarse_hyperparameters_data.sort_values(by="score")
    
    coarse_hyperparameters_data.to_csv(filename)
    
coarse_hyperparameters_data.head(10)
```

```python
#Bayesian Optimization을 통한 hyperparameter tuning 준비과정
from sklearn.preprocessing import LabelEncoder
# creating instance of labelencoder
labelencoder = LabelEncoder()
# Assigning numerical values and storing in another column
y_train = labelencoder.fit_transform(y_train)
```

```python
import warnings
warnings.filterwarnings("ignore")

from bayes_opt import BayesianOptimization
import lightgbm as lgb

def bayes_parameter_opt_lgb(X, y, init_round=15, opt_roun=25, n_folds=7, random_seed=42, n_estimators=10, learning_rate=0.02, output_process=False,colsample_bytree=0.93,min_child_samples=56,subsample=0.84,min_data_in_leaf=1000):
    # prepare data
    train_data = lgb.Dataset(data=X_train, label=y_train)
    # parameters
    def lgb_eval(num_leaves, feature_fraction, bagging_fraction, max_depth, reg_alpha, reg_lambda, min_split_gain, min_child_weight, colsample_bytree,min_child_samples,subsample,learning_rate,min_data_in_leaf):
        params = {'application':'multiclass','num_iterations': n_estimators, 'learning_rate':learning_rate, 'early_stopping_round':100, 'metric':'multi_logloss'}
        params['learning_rate']
        params["num_leaves"] = int(round(num_leaves))
        params["num_class"] = 39
        params['feature_fraction'] = max(min(feature_fraction, 1), 0)
        params['bagging_fraction'] = max(min(bagging_fraction, 1), 0)
        params['max_depth'] = int(round(max_depth))
        params['reg_alpha'] = max(reg_alpha, 0)
        params['reg_lambda'] = max(reg_lambda, 0)
        params['min_split_gain'] = min_split_gain
        params['min_child_weight'] = min_child_weight
        params['colsample_bytree'] = colsample_bytree
        params['min_child_samples'] = min_child_samples,
        params['subsample'] = 0.84,
        params['min_data_in_leaf'] = int(min_data_in_leaf)
        cv_result = lgb.cv(params, train_data, nfold=n_folds, seed=random_seed, stratified=True, verbose_eval =200, metrics=['multi_logloss'])
        return max(cv_result['multi_logloss-mean'])
    # range 
    lgbBO = BayesianOptimization(lgb_eval, {'num_leaves': (10, 200),
                                            'learning_rate':(0.001, 0.1),
                                            'feature_fraction': (0.1, 0.9),
                                            'bagging_fraction': (0.8, 1),
                                            'max_depth': (5, 8.99),
                                            'reg_alpha': (1, 10),
                                            'reg_lambda': (1, 15),
                                            'min_split_gain': (0.001, 0.1),
                                            'min_child_weight': (5, 50),
                                            'colsample_bytree' : (0.7,1.0),
                                            'min_child_samples' : (2,200),
                                            'min_data_in_leaf': (10, 200), 
                                            'subsample' : (0.7,1.0)
                                           }, random_state=0)
    # optimize
    lgbBO.maximize(init_points=init_round, n_iter=opt_roun)
    
    # output optimization process
    if output_process==True: lgbBO.points_to_csv("bayes_opt_result.csv")
    
    # return best parameters
    return lgbBO

opt_params = bayes_parameter_opt_lgb(X_train, y_train, init_round=10, opt_roun=10, n_folds=10, random_seed=42, n_estimators=10)
```

```python
params = opt_params.res[16]['params']
params['max_depth'] = int(round(params['max_depth']))
params['num_leaves'] = int(round(params['num_leaves']))
params['min_data_in_leaf'] = int(round(params['min_data_in_leaf']))

params
```

```python
# 튜닝 종료 된 패러미터로 모델 재설정 
lgbm_model.set_params(**params)
```

```python
# Validation 데이터셋으로 검정 실시
lgbm_model.fit(X_train_kf,y_train_kf)
```
